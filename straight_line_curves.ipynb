```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

num_lines = 5
m1, c1 = -1, 10
```

```python
t1 = np.arange(0, 10.1, 0.5)
df_data = pd.DataFrame(index=t1)
def f(x,m,c):
    return m*x+c

df_data['y1'] = f(t1,-4,4)
df_data['y2'] = f(t1,-3/2,3)
df_data['y3'] = f(t1,-2/3,2)
df_data['y4'] = f(t1,-1/4,1)

df_data[df_data < 0] = np.nan

plt.plot(df_data, c='k')
plt.show()
```

```python

t1 = np.arange(0, num_lines*1.01, 0.5)
df_data = pd.DataFrame(index=t1)

def f(x,m,c):
    return m*x+c

df_data['y1'] = f(t1,-4,4)
df_data['y2'] = f(t1,-3/2,3)
df_data['y3'] = f(t1,-2/3,2)
df_data['y4'] = f(t1,-1/4,1)

df_data[df_data < 0] = np.nan

plt.plot(df_data, c='k')
plt.show()
```

```python
num_lines = 4
t1 = np.arange(0, num_lines*1.01, 0.5)
data = []
for line in range(num_lines):
    print(f"Line: {line+1}")
    c = num_lines-line
    print(f"  c: {c}, m: {-c/(line+1)}")
    print(f"  y= {f(t1, -c/(line+1), c)}")
    data.append(f(t1, -c/(line+1), c))

df = pd.DataFrame(data).T
df[df < 0] = np.nan

plt.plot(
    df,
    c='k',
    linewidth=0.6
    )
#plt.axis([0, num_lines, 0, num_lines])
plt.show()
```
